```python
import pandas as pd
```

```python
hi_file = "../Data/health_insurance(2017-2018).csv"
hi_df = pd.read_csv(hi_file)
hi_df
```

```python
hi_2018_df = hi_df.drop_duplicates("Name", keep="first")
hi_2018_df = hi_2018_df.reset_index(drop=True)
hi_2018_df
```

```python
hi_2017_df = hi_df.drop_duplicates("Name", keep="last")
hi_2017_df = hi_2017_df.reset_index(drop=True)
hi_2017_df.head()
```

```python
hi_2018_reduced_df = hi_2018_df[["Name","Uninsured: %"]]
hi_2017_reduced_df = hi_2017_df[["Name","Uninsured: %"]]
hi_combined_df = pd.merge(hi_2017_reduced_df,hi_2018_reduced_df, on="Name", suffixes=(' 2017',' 2018'))
hi_combined_df.head()
```

```python
hi_final_df = hi_combined_df.rename(columns={"Name":"State","Uninsured:% 2017":"2017_uninsured(%)","Uninsured:% 2018":"2018_uninsured(%)"})
hi_final_df.head()
```
